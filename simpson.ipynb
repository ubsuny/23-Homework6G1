```python
import calculus as calc
import matplotlib.pyplot as plt
import numpy as np
```

```python
GROUP_NUMBER = 1 # specifing our group number

# defining our functions
def exp_func(x):
  return np.exp(-1 / np.where(np.isclose(x, 0), 1e-8, x))

def cos_func(x):
  return np.cos(1 / np.where(np.isclose(x, 0), 1e-8, x))

def cube_func(x):
  return x**3 + 1 / GROUP_NUMBER
```

```python
# integral of exp(-1/x)

# plotting graph using first set of bounds
lower_bound = 0
upper_bound = 10*GROUP_NUMBER
x_vals = np.linspace(lower_bound, upper_bound, 100)
y_vals = exp_func(x_vals)
plt.plot(x_vals,y_vals)
plt.fill_between(x_vals, y_vals, alpha = 0.5)
plt.title("Integral of e^(-1/x) (0 to 10)")
plt.xlabel('x values')
plt.ylabel('y values')
```

```python
calc.simpson(exp_func, lower_bound, upper_bound,10) # value of integral from 0 to 10
```

```python
# plotting graph using second set of bounds
lower_bound = 0
upper_bound = np.pi*GROUP_NUMBER
x_vals = np.linspace(lower_bound, upper_bound, 100)
y_vals = exp_func(x_vals)
plt.plot(x_vals,y_vals)
plt.fill_between(x_vals, y_vals, alpha = 0.5)
plt.title("Integral of e^(-1/x)  (0 to pi)")
plt.xlabel('x values')
plt.ylabel('y values')
```

```python
calc.simpson(exp_func, lower_bound, upper_bound,10) # value of integral from 0 to pi
```

```python
# plotting graph using third set of bounds
lower_bound = -1
upper_bound = 1
x_vals = np.linspace(lower_bound, upper_bound, 100)
y_vals = exp_func(x_vals)
plt.plot(x_vals,y_vals)
plt.fill_between(x_vals, y_vals, alpha = 0.5)
plt.title("Integral of e^(-1/x)  (-1 to 1)")
plt.xlabel('x values')
plt.ylabel('y values')
```

```python
calc.simpson(exp_func, lower_bound, upper_bound,10) # value of integral from -1 to 1
```

```python
# integral of cos(1/x)
# plotting graph using first set of bounds
lower_bound = 0
upper_bound = 10*GROUP_NUMBER
x_vals = np.linspace(lower_bound, upper_bound, 100)
y_vals = cos_func(x_vals)
plt.plot(x_vals,y_vals)
plt.fill_between(x_vals, y_vals, alpha = 0.5)
plt.title("Integral of cos(1/x) (0 to 10)")
plt.xlabel('x values')
plt.ylabel('y values')
```

```python
calc.simpson(cos_func, lower_bound, upper_bound,10) # value of integral from 0 to 10 for cos function
```

```python
# plotting cos(1/x) using second set of bounds
lower_bound = 0
upper_bound = np.pi*GROUP_NUMBER
x_vals = np.linspace(lower_bound, upper_bound, 100)
y_vals = cos_func(x_vals)
plt.plot(x_vals,y_vals)
plt.fill_between(x_vals, y_vals, alpha = 0.5)
plt.title("Integral of cos(1/x)  (0 to pi)")
plt.xlabel('x values')
plt.ylabel('y values')
```

```python
calc.simpson(cos_func, lower_bound, upper_bound,10) # value of integral from 0 to pi for cos function
```
